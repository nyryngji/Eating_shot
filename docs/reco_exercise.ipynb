```python
import pandas as pd 
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import random

exercise_dataset = pd.read_csv('exercise_category.csv')
exercise_dataset.head(20)

# location:1이면 실내, type:1이면 유산소
```

```python
exercise_dataset['location'] = np.where(exercise_dataset['location'] == '실외', 0, 1)
exercise_dataset['type'] = np.where(exercise_dataset['type'] == '유산소', 1, 0)
```

```python
def exercise_recommend(user_input):
    recommend_list = pd.DataFrame()

    more_ex = user_input[2:].index(min(user_input[2:])) if abs(user_input[2] - user_input[3]) >= 1 else 0
    # 만약 유산소와 무산소 운동 횟수가 차이가 난다면 더 적게 한 운동을 추천해주도록 유도, 아니면 무산소 운동 추천
    # 상대적으로 무산소 운동이 항목이 적어서 이러한 조건문 사용

    if user_input[1] >= 18 or user_input[1] <= 5:
        # 만약 늦은 저녁이나 새벽이면 실내 운동을 추천
        recommend_list = exercise_dataset[exercise_dataset['운동 종류'].isin(exercise_dataset[exercise_dataset['location']==1]['운동 종류'])]
    else:
        recommend_list = exercise_dataset[exercise_dataset['type'] == abs(1-more_ex)]
                    
    similarity_scores = cosine_similarity(recommend_list[recommend_list.columns[1:3]], [user_input[1:3]])

    # 원래 가장 높은 것을 추천하는 코드를 작성하였으나 계속 같은 종류의 운동이 나오기에 상위 5개의 운동 중 랜덤 선정으로 바꿈
    return recommend_list['운동 종류'][:5].iloc[random.randint(0,4)]
```

```python
exercise_recommend([800, 10, 3, 2])
# user_input = [1000-현재 운동한 칼로리, 현재 시간, 유산소 운동 횟수, 무산소 운동 횟수]
```
