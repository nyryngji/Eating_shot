```python
import random

import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

exercise_dataset = pd.read_csv("exercise_category.csv")
exercise_dataset[exercise_dataset["location"] == "실내"]
# location:1이면 실내, type:1이면 유산소
```

```python
exercise_dataset["location"] = np.where(exercise_dataset["location"] == "실외", 0, 1)
exercise_dataset["type"] = np.where(exercise_dataset["type"] == "유산소", 1, 0)
```

```python
def exercise_recommend(user_input):
    recommend_list = pd.DataFrame()

    more_ex = (
        user_input[2:].index(min(user_input[2:]))
        if abs(user_input[2] - user_input[3]) >= 1
        else 0
    )

    if user_input[1] >= 18 or user_input[1] <= 5:
        recommend_list = exercise_dataset[
            exercise_dataset["운동 종류"].isin(
                exercise_dataset[exercise_dataset["location"] == 1]["운동 종류"]
            )
        ]
    else:
        recommend_list = exercise_dataset[exercise_dataset["type"] == abs(1 - more_ex)]

    similarity_scores = cosine_similarity(
        recommend_list[recommend_list.columns[1:3]], [user_input[1:3]]
    )

    print(recommend_list)

    return recommend_list["운동 종류"][:5].iloc[random.randint(0, 4)]


user_input = [790, 19, 1, 0]
print(exercise_recommend(user_input))
```
